```python
print("Ejercicio Número 1")
# Sin utilizar el comando abs, escribir un programa que calcule e imprima \
# el valor absoluto de cualquier número (sea entero o decimal).

print("Este programa calcula el valor absoluto de cualquier número.")

Num = (float(input("ingrese un número: ")))  #Ingresamos el número.

if Num >= 0:    # Si es mayor o igual a "cero" se imprime tal cual.            
    print(f"El valor absoluto de {Num} es: ",Num)
    
elif Num <  0:    #Si es menor que "cero" se multiplica por "menos uno". 
    AbsNum = Num * -1    #El valor positivo se guarda en una nueva variable.
    print(f"El valor absoluto de {Num} es: ",AbsNum) 
```

```python
# Leer dos números enteros y determinar si la magnitud de la diferencia \
# entre los dos es un número primo. 

print("Ejecicio número 2")
PriNum = (int(input("Por favor ingrese el primer número entero: ")))

SegNum = (int(input("Por favor ingrese el segundo número entero: ")))

MagDif = abs(PriNum - SegNum)  #Se obtiene el valor absoluto de la diferencia.

for i in range(2,MagDif+1):  #i recorre desde 2 hasta el valor de MagDif
    a = MagDif % i    #Se guarda en a el residuo de la división MagDif/i
    b = MagDif / i   #Se guarda en b el cociente de la división MagDif/i
    
    if a == 0:   #Si el residuo da cero en algún momento del ciclo no es primo. 
        print(f"{MagDif} no es primo.")
        break   #Cancela el ciclo
    elif b < i: #Si el cociente es menor al divisor en algún momento del ciclo\
                 # MagDif es primo.
        print(f"{MagDif} es primo.")
        break    #Cancela el ciclo
```

```python
# Iterar a través de los primeros cien enteros positivos, buscando los \
# múltiplos de 3 e imprimiendo y almacenándolos en una lista hasta encontrar \
# los primeros 15 de ellos. Una vez encontrados, continuar iterando en busca \
# de los múltiplos de 4 y almacenarlos en otra lista. 

print("Ejercicio número 3")

Multiplos_3 = []  #Lista para los múltiplo de 3
Multiplos_4 = []  #Lista para los múltiplos de 4

Num_Multi = 0   

for j in range(1,101):
    
    while len(Multiplos_3) <= 14:  #Funciona mientras Num_Multi sea <= a 14
        a = j % 3                   #Condición para ser multiplo de 3
        if a == 0:
            Multiplos_3.append(j) #Agrega los valores j a la lista Multiplo_3
           
        Num_Multi = Num_Multi + 1
        b = j          # Guarda en b el último valor de j agaregado a la lista
        break          #Cierra el ciclo
        


for i in range(b,101):
    c = i%4                      #Condición para ser multiplo de 4
    if c == 0:
        Multiplos_4.append(i)   #Agrega los valores i a la lista Multiplo_4
 
print(f"Los primeros 15 múltiplos de 3 son {Multiplos_3}")                  
print(f"Los multiplos de 4 después del 15to múltiplo de 3 son {Multiplos_4}")
```

```python
# Hacer un programa que lea las coordenadas (x1,y1,r1) y (x2,y2,r2) que 
#corresponden al centro y al radio de dos círculos. Lea un punto de 
#coordenadas (a,b). Determinar si (a,b) está contenido: a) dentro del círculo 1
# b) dentro del círculo 2; c) dentro de ambos círculos;
# d) fuera de ambos círculos. Recordemos que el círculo es el lugar geométrico
# de los puntos del plano cuya distancia a otro punto fijo, llamado centro, 
#es menor o igual que una cantidad constante, llamada radio. 

import math
print("Ejercicio npumero 4")

Cor_x1=(float(input("Ingrese la coord x del centro del primer círculo: ")))
Cor_y1=(float(input("Ingrese la coord y del centro del primer círculo: ")))
Radio1 = abs((float(input("Ingrese el valor del radio del primer círculo:"))))

Cor_x2=(float(input("Ingrese la coord x del centro del segundo círculo: ")))
Cor_y2=(float(input("Ingrese la coord y del centro del segundo círculo: ")))
Radio2 =abs((float(input("Ingrese el valor del radio del segundo círculo: "))))

x3= (float(input("Ingrese el valor de la coordenada x del punto: ")))
y3= (float(input("Ingrese el valor de la coordenada y del punto: ")))


hip1 = math.sqrt((x3-Cor_x1)**2 + (y3-Cor_y1)**2)  #Guarda el valor de la 
#hipotenusa entre el punto ingresado y el centro del primer circulo

if hip1 <= Radio1:  #Condición para que una coordenada esté dentro de un 
    #circulo
    print(f"El punto({x3} , {y3}) Si està dentro del primer cìrculo.") 
elif hip1 > Radio1:   #Condición para que una coordenada esté fuera de un 
    #circulo
    print(f"El punto({x3} , {y3}) No està dentro del primer cìrculo.") 

hip2 = math.sqrt((x3-Cor_x2)**2 + (y3-Cor_y2)**2)  #Guarda el valor de la 
#hipotenusa entre el punto ingresado y el centro del segundo circulo

if hip2 <= Radio2:  #Condición para que una coordenada esté dentro de un 
    #circulo
    print(f"El punto({x3} , {y3}) Si està dentro del segundo cìrculo.")
elif hip2 > Radio2:  #Condición para que una coordenada esté fuera de un 
    #circulo
    print(f"El punto({x3} , {y3}) No està dentro del segundo cìrculo.")

if hip1 <= Radio1 and hip2<= Radio2: #Condición para que una coordenada 
    #esté dentro de ambso círculos circulo
    print(f"El punto ({x3},{y3}) Si está dentro de ambos círculos.")

elif hip1 >= Radio1 and hip2 >= Radio2: #Condición para que una coordenada 
    #esté fuera de ambso círculos circulo
    print(f"El punto ({x3},{y3}) No está dentro de ambos círculos.")
```

```python
print("Ejercicio npumero 5")
cadena = input("Ingrese una cadena de texto: ")  #Guarda la cadena ingresada

Vocales_Mayus = "AEIOU"    #guarda como cadena en una variable las vocales 
#mayusculas
Num_vocales = 0            #contador de vocales matusculas

letras_tilde = "áéíóúÁÉÍÓÚ"   #guarda como cadena en una variable las 
#letras que puden tener tilde
Num_letras_tilde = 0    #contador de letras con tilde

Digitos = "123456789"    #guarda como cadena los digitos
Num_digitos = 0     #contador de digitos

Espacios = " "    #guarda como cadena el espacio entre palabras
Num_espacios = 0   #contador de espacios

Reservadas = ["False","True","None","raise","assert","not","class","pass",\
            "continue","global","def","del","elif","else","finally","for",\
            "from","if","import","in","is","lambda","except","nonlocal",\
            "or","break","return","try","and","while","with","as","yield"]
#Guarda en una lista las palbras reservadas más usadas en este lenguaje
Num_reservadas = 0   #contador de palabras reservdas

for j in range(len(cadena)):
    if cadena[j] in Vocales_Mayus:  #itera en la variable cadena comparando 
        #con la cadena Vocales_Mayus 
        Num_vocales += 1  #Si coinciden los elementos se agrega 1 al 
        #contador respectivo
        
    elif cadena[j] in letras_tilde:  #itera en la variable cadena comparando 
        #con la cadena letras_tilde
        Num_letras_tilde += 1  #Si coinciden los elementos se suma 1 al 
        #contador respectivo
        
    elif cadena[j] in Digitos:  #itera en la variable cadena comparando 
        #con la cadena Digitos
        Num_digitos += 1  #Si coinciden los elementos se suma 1 al 
        #contador respectivo
        
    elif cadena[j] == Espacios:  #itera en la variable cadena comparando 
        #con la cadena Espacios
        Num_espacios += 1  #Si coinciden los elementos se suma 1 al 
        #contador respectivo
cadena_lista = cadena.split()  #convierte la variable cadena de str a list 
 
for n, i in enumerate(cadena_lista):  #itera en la variable cadena comparando 
        #con la cadena Reservadas
    if i in Reservadas:      
        Num_reservadas += 1   #Si coinciden los elementos se suma 1 al 
        #contador respectivo
        
print(f'Se entraron {Num_vocales} letras vocales en mayúscula')
print(f'Se entraron {Num_letras_tilde} letras con tilde')
print(f'Se entraron {Num_digitos} digitos')
print(f'Se entraron {Num_espacios} espacios')
print(f'Se entraron {Num_reservadas} palabras reservadas')
       
```

```python
print("Ejercicio npumero 8")
lista_num=[]                                           #Se crea la lista vacía
long_lista = int(input("Digite el tamaño de la lista: ")) #Tamaño de la lista

for elemento in range(long_lista):
    lista_num.append(int(input(f"Digite el número {elemento+1}: ")))#Crea lista
print("La lista es: ",lista_num)

lista=lista_num.copy()     #Se copia la lista generada en una nueva variable
maximo = max(lista)        #Guarda en maximo el valor maximo de la lista
a = lista.count(maximo)  # Guarda en a el número de veces q se repite el maximo                
c = len(lista)            #Guarda en c la longitud de la lista

if a == c:   # Si el maximo se repite tantas veces como la longitud entonces:
    print("No hay un segundo elemento mayor")
      

else:           #De lo contrario:
    for i in range(a):   #Con i variando en el rango de repeticiones:
        lista.remove(max(lista))  #se remueve i veces el máximo
        newlista = lista.copy()   #se crea la nueva lista sin el máximo
        sec_mayor = max(newlista)    #se obtiene el nuevo máximo     
    print(f"El segundo elemento más grande de la lista es: ",sec_mayor)
```

```python
print("Ejercicio Número 10")

Matriz=[]   #Matriz que guardara los datos ingresados por el usuario
Matriz_desenredada = []

filas = int(input('Digite el # de filas de la matriz: ')) 
columnas=int(input('Digite el # de columnas de la matriz: '))

for i in range(filas):
    Matriz.append([])
    for j in range(columnas):
        Matriz[i].append(int(input(f'Escriba el elemento #{j+1} de la {i+1}\
 fila: ')))
#Con el ciclo anterior se creo la matriz, iterando en las filas y columnas 
#ingresadas        
print('\nLa matriz ingresada es: ',Matriz)            


elem = filas *  columnas
columna_final = columnas -1
columna_inicial = 0
fila_final = filas -1
fila_inicial = 0
lim = 0

while lim < elem: #Funciona hasta que se hallan extraido todos los elementos
    
    for i in range(columna_final,columna_inicial,-1): #primer recorrido,
        #en la ultima fila de derecha a izquierda.
        pos = Matriz[fila_final][i] #Extrae el elemento
        Matriz_desenredada.append(pos)#Se agrega al vector el dato extraido
        lim +=1 #Cuenta la cantidad de elementos extraídos
        if lim == elem:
            break
   
    for i in range(fila_final,fila_inicial,-1):#Para el segundo recorrido
        pos = Matriz[i][columna_inicial]#Se extrae el elemento de la matriz
        Matriz_desenredada.append(pos)#Se agrega al vector el dato extraido
        lim +=1 #Contador para la cantidad de elementos extraídos
        if lim == elem:
            break
    for i in range(columna_inicial,columna_final):#Para el tercer recorrido
        pos = Matriz[fila_inicial][i]#Se extrae el elemento de la matriz
        Matriz_desenredada.append(pos)#Se agrega al vector el dato extraido
        lim +=1 #Contador para la cantidad de elementos extraídos
        if lim == elem:
            break
    for i in range(fila_inicial,fila_final):#Para el cuarto recorrido
        pos = Matriz[i][columna_final]#Se extrae el elemento de la matriz
        Matriz_desenredada.append(pos)#Se agrega al vector el dato extraido
        lim +=1 #Contador para la cantidad de elementos extraídos
        if lim == elem:
            break
    columna_final -= 1
    columna_inicial += 1
    fila_final -= 1
    fila_inicial += 1
    #Se ajusta las posiciones de los proximos recorridos de los ciclos
print("El vector pedido es: ",Matriz_desenredada) #Imprime
```

```python
import random as ran    # librería para generar números aleatorios
print("Ejercicio número 14")
def suma():
    ''' Esta función me permite generar sumas y ver si el valor ingresado es\
    correcto'''
    aciertos = 5     # número de veces que se tiene que acertar en la suma
    
    while aciertos > 0:
        var1 = ran.randint(1,100)    #primer número generado
        var2 = ran.randint(1,100)    # segundo número generado
        resultado_suma = var1 + var2  # valor de la suma propuesta
        print(f"La suma propuesta es {var1} + {var2}") #muestra la suma
        d = float(input("Digite el valor de la suma :"))
        if d == resultado_suma:
            print("Es correcto, la suma es :",resultado_suma)
            aciertos = aciertos -1  #la variable aciertos disminuye hasta cero
            f = 5 - aciertos        # total de aciertos
            print(f"Ha tenido {f} sumas correctas")
        else:
                print("Es incorrecto, vuelva a intentarlo")
                
    print("Felicitaciones, ha terminado.")  # el programa ha terminado
suma()
```
